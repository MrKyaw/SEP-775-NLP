```python

import pandas as pd
from datetime import datetime, timedelta
import random

# Generate sample data
dates = [datetime.now() - timedelta(days=i) for i in range(30)]
products = ['Product A', 'Product B', 'Product C']
data = {
    'Date': [date.strftime('%Y-%m-%d') for date in dates for _ in products],
    'Product': products * len(dates),
    'Demand': [random.randint(50, 200) for _ in range(len(dates) * len(products))]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Save to CSV
df.to_csv('demand_history.csv', index=False)

print("demand_history.csv created successfully.")
```
